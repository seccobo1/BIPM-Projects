## 2.1 Finding Missing Value

```python
#import all libraries
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load in data
df = pd.read_json('data/train.json')
df.head(5)
```

```python
# find the missing data for each variable

# bathrooms

bath_series = df[df['bathrooms'] == 0]
bath = bath_series['bathrooms']
bath_m = bath.count()

# bedrooms

bed_series = df[df['bedrooms'] == 0]
bed = bed_series['bedrooms']
bed_m = bed.count()

# building id

bid_series = df[df['building_id'] == '0']
bid = bid_series['building_id']
bid_m = bid.count()

# created (date)

df['created'] = pd.to_datetime(df['created'])
created = df['created']
created_m = 0 # no missing data

# description

desc_series = df['description'].str.strip()
desc_count1 = desc_series.where(desc_series=='').count()
desc_count2 = desc_series.where(desc_series=='<p><a  website_redacted').count()
desc_m = desc_count1+desc_count2

# display address

d_add = df['display_address'].str.strip()
d_add_m = d_add.where(d_add=='').count()

# features

features_series = df['features'].str.len()
feat_m = features_series[features_series == 0].count()

# interest level

intl_series = df['interest_level']
filt_low = intl_series == 'low'
filt_med = intl_series == 'medium'
filt_high = intl_series == 'high'
intl_m = intl_series.where(~(filt_low | filt_med | filt_high)).dropna().count()

# latitude

lat_series = df[df['latitude'] == 0]
lat = lat_series['latitude']
lat_m = lat.count()

# listing id

lid_series = df['listing_id']

# longitude

lon_series = df[df['longitude'] == 0]
lon = lon_series['longitude']
lon_m = lon.count()

# manager id

mid = df['manager_id'].str.strip()
#mid.where(mid==0).dropna()
mid.where(mid=='').dropna()
mid_m = mid.count()

# photos

photo_series = df['photos'].str.len()
photos = df['photos']
p_m = photo_series[photo_series == 0].count()

# price

price_series = df[df['price'] == 0]
price = price_series['price']

# street address

s_add = df['street_address'].str.strip()
s_m = s_add.where(s_add=='').count()
```

### 2.2.2 Identify outliers for the "Bathrooms" field

After visiting the renthop website, we can assume that for "Studio" room, the number of bedrooms seems to set to 0 in our dataset. However, from the bathrooms filter as we can see from the picture above, the number of bathrooms must be 1+bath.

Therefore, the bathroom value of 0 is invalid.

```python
df.info()
```

Here, we can see that we have 49352 entries.

```python
new_df = df.loc[df['bathrooms'] > 0]
indexes = list(new_df.index.values.tolist())
values = list(new_df['bathrooms'][indexes])
```

```python
plt.scatter(indexes, new_df['bathrooms'][indexes], edgecolor = 'black')
plt.xlabel('indexes')
plt.ylabel('Number of Bathrooms')
plt.show()
new_df['bathrooms'].describe()
```

From previous info we know that the dataset **df** contains 49352 entries. and after removing 0 bathroom, the remaining data is 49039 entries.

From the scaterplot graph above, we can see the is a data at 10 bathroom, is it an outlier?

```python
new_df[new_df['bathrooms'] == 10].describe()
```

```python
new_df[new_df['bathrooms'] == 5].describe()
```

We can see that the mean price of listing with "10 bathrooms" is a lot cheaper that the minimum listing with "5 bathrooms". Therefore, we assume "10 bathrooms" is an outlier.

```python
x = df['bathrooms'].loc[(df['bathrooms'] == 0) | (df['bathrooms'] == 10)].count()
print("Number of outliers:", x)
```

### 2.2.3 Identify outliers for the "bedrooms" field

As mention above at point 1.2.2, we assume that selecting "Studio" will set the number of bedrooms to 0. So, we keep rows that contains 0 bedroom. Furthermore, as mention in our EDA while comparing number of bedrooms and their price, the mean of their price does not make sense. Therefore, we are define it as outliers.

```python
indexes = list(df.index.values.tolist())
    
plt.scatter(indexes, df['bedrooms'][indexes], edgecolor = 'black')
plt.xlabel('indexes')
plt.ylabel('Number of Bedrooms')
plt.show()
df['bedrooms'].describe()
```

```python
sns.boxplot(x=df['bedrooms'])
```

```python
#new_df = df.loc[df['bedrooms'] > 3]
new_df = df.loc[(df['bedrooms'] > 3) & (df['bedrooms'] < 7)]
sns.boxplot(x=new_df['bedrooms'])
plt.show()
```

We can see that most of our data lies between 1 and 2. Other bedroom between 4 and 6 maybe outliers as they are far from the quartiles. So, we invetigate those points by ploting other graph to see their quartiles, which only shows a single vertical line, it means they don't have relevance to each other. However, we don't know if they are meaningful - they are factual/valid data.

Lets compare the bedroom between 4 and 6 with their prices.

```python
df['price'].loc[df['bedrooms'] > 3].describe()
```

```python
df['price'].loc[df['bedrooms'] <= 3].describe()
```

The mean price of listings with the number of bedrooms between 4 and 6 is greater than the mean price of listings with the number of bedrooms below 3. This assures that all "bedrooms" values are reasonable, therefore we should keep bedroom between 4 and 6 as well.

## 2.3 Drop outliers

With the following code, we will now clean the data according to the beforementioned findings. This code will be from now an applied in the beginning of every notebook, to make sure the dataframe is cleaned before working with it.

```python
df = df[(df['bathrooms'] != 0) | (df['bathrooms'] != 10)]
df = df[df['bedrooms'] <= 6]
df = df[df['building_id'].str.len() >= 32]
df = df[df['price'] < 1000000]
df.head()
```
